## 1. Variables and Data Types

### What is a Variable?
A variable is like a labeled box that stores information. You give it a name and put data inside.

### Python's Main Data Types:
- **int**: Whole numbers (e.g., 42, -5)
- **float**: Decimal numbers (e.g., 3.14, -0.5)
- **str**: Text (e.g., "hello", 'world')
- **bool**: True or False

Let's try it:

```python
# Creating variables
energy_consumption = 1500  # int (whole number)
temperature = 23.5  # float (decimal)
building_name = "Main Office"  # string (text)
is_operational = True  # boolean (True/False)

# Let's see what we stored
print("Energy Consumption:", energy_consumption)
print("Temperature:", temperature)
print("Building Name:", building_name)
print("Is Operational:", is_operational)

# Check the type of a variable
print("\nType of energy_consumption:", type(energy_consumption))
print("Type of temperature:", type(temperature))
```

## 2. Basic Operations

### Math Operations

```python
# Math operations
a = 10
b = 3

print("Addition:", a + b)  # 13
print("Subtraction:", a - b)  # 7
print("Multiplication:", a * b)  # 30
print("Division:", a / b)  # 3.333...
print("Floor Division:", a // b)  # 3 (rounds down)
print("Modulus (remainder):", a % b)  # 1
print("Power:", a ** b)  # 1000 (10 to the power of 3)
```

### String Operations

```python
# String operations
first_name = "John"
last_name = "Doe"

# Concatenation (joining strings)
full_name = first_name + " " + last_name
print("Full Name:", full_name)

# String formatting (f-strings)
age = 25
message = f"My name is {full_name} and I am {age} years old."
print(message)

# Useful string methods
text = "  Energy Anomaly Detection  "
print("Original:", text)
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
print("Stripped:", text.strip())  # Removes whitespace
print("Length:", len(text))
```

## 3. Lists - Storing Multiple Items

A **list** is a collection of items in order. Think of it like a shopping list.

```python
# Creating a list
energy_readings = [1450, 1520, 1380, 1600, 1490]
print("Energy Readings:", energy_readings)

# Accessing items (indexing starts at 0)
print("First reading:", energy_readings[0])  # 1450
print("Third reading:", energy_readings[2])  # 1380
print("Last reading:", energy_readings[-1])  # 1490

# List operations
energy_readings.append(1550)  # Add to end
print("After append:", energy_readings)

# List slicing
print("First 3 readings:", energy_readings[:3])
print("Last 2 readings:", energy_readings[-2:])

# List methods
print("Length:", len(energy_readings))
print("Maximum:", max(energy_readings))
print("Minimum:", min(energy_readings))
print("Average:", sum(energy_readings) / len(energy_readings))
```

## 4. Dictionaries - Key-Value Pairs

A **dictionary** stores data in key-value pairs. Like a real dictionary where you look up a word (key) to get its definition (value).

```python
# Creating a dictionary
building = {
    "name": "Main Office",
    "floors": 5,
    "area_sqm": 2500,
    "is_smart": True,
    "sensors": ["temperature", "humidity", "energy"]
}

# Accessing values
print("Building Name:", building["name"])
print("Number of Floors:", building["floors"])

# Adding new key-value pairs
building["year_built"] = 2015
print("\nUpdated building:")
print(building)

# Dictionary methods
print("\nAll keys:", building.keys())
print("All values:", building.values())
print("All items:", building.items())
```

## 5. Control Flow - Making Decisions

### if/elif/else Statements

```python
# Simple if statement
temperature = 28

if temperature > 25:
    print("It's hot! Turn on the AC.")
elif temperature > 18:
    print("Temperature is comfortable.")
else:
    print("It's cold! Turn on the heater.")

# Checking anomalies
energy_consumption = 2500
normal_range = (1000, 2000)

if energy_consumption < normal_range[0]:
    print("⚠️ Unusually LOW energy consumption detected!")
elif energy_consumption > normal_range[1]:
    print("⚠️ Unusually HIGH energy consumption detected!")
else:
    print("✅ Energy consumption is normal.")
```

## 6. Loops - Repeating Actions

### For Loops

```python
# Loop through a list
energy_readings = [1450, 1520, 1380, 1600, 1490]

print("Checking each reading:")
for reading in energy_readings:
    if reading > 1500:
        print(f"  {reading} kWh - HIGH")
    else:
        print(f"  {reading} kWh - Normal")

# Loop with index
print("\nWith index:")
for i, reading in enumerate(energy_readings):
    print(f"  Reading {i+1}: {reading} kWh")
```

### While Loops

```python
# While loop example
count = 0
total = 0

print("Counting energy readings:")
while count < 5:
    total += energy_readings[count]
    count += 1
    print(f"  Added reading {count}: Total = {total}")

print(f"\nAverage energy: {total / count} kWh")
```

## 7. Functions - Reusable Code Blocks

Functions help you organize code and avoid repetition.

```python
# Simple function
def greet(name):
    """Greet someone by name"""
    return f"Hello, {name}!"

message = greet("Alice")
print(message)

# Function with multiple parameters
def calculate_cost(energy_kwh, price_per_kwh=0.12):
    """
    Calculate energy cost
    
    Args:
        energy_kwh (float): Energy consumed in kWh
        price_per_kwh (float): Price per kWh (default: $0.12)
    
    Returns:
        float: Total cost
    """
    cost = energy_kwh * price_per_kwh
    return cost

# Using the function
monthly_energy = 1500
cost = calculate_cost(monthly_energy)
print(f"Energy cost for {monthly_energy} kWh: ${cost:.2f}")

# Using custom price
cost_expensive = calculate_cost(monthly_energy, price_per_kwh=0.18)
print(f"With expensive rate: ${cost_expensive:.2f}")
```

### Real-World Example: Anomaly Detection Function

```python
def detect_anomaly(value, normal_min, normal_max):
    """
    Detect if a value is outside normal range
    
    Args:
        value (float): The value to check
        normal_min (float): Minimum normal value
        normal_max (float): Maximum normal value
    
    Returns:
        dict: Anomaly status and type
    """
    if value < normal_min:
        return {"is_anomaly": True, "type": "LOW", "severity": normal_min - value}
    elif value > normal_max:
        return {"is_anomaly": True, "type": "HIGH", "severity": value - normal_max}
    else:
        return {"is_anomaly": False, "type": "NORMAL", "severity": 0}

# Test the function
readings = [1200, 1500, 2500, 900, 1600]

print("Anomaly Detection Results:")
for reading in readings:
    result = detect_anomaly(reading, 1000, 2000)
    status = "⚠️  ANOMALY" if result["is_anomaly"] else "✅ Normal"
    print(f"  {reading} kWh: {status} - {result['type']}")
```

### 📝 Your Turn:
Create a function that:
- Takes a list of temperatures
- Calculates the average temperature
- Returns "High" if avg > 25, "Normal" if 18-25, "Low" if < 18

```python
def analyze_temperature(temperatures):
    """Analyze a list of temperature readings"""
    # Calculate average
    avg_temp = sum(temperatures) / len(temperatures)
    
    # Determine category
    if avg_temp > 25:
        category = "High"
    elif avg_temp >= 18:
        category = "Normal"
    else:
        category = "Low"
    
    return f"Average: {avg_temp:.1f}°C - Category: {category}"

# Test it
temps = [22, 24, 23, 25, 21]
result = analyze_temperature(temps)
print(result)
```

## 8. List Comprehensions - Elegant Loops

A compact way to create lists from other lists.

```python
# Traditional way
numbers = [1, 2, 3, 4, 5]
squared = []
for num in numbers:
    squared.append(num ** 2)
print("Traditional:", squared)

# List comprehension way (much cleaner!)
squared = [num ** 2 for num in numbers]
print("Comprehension:", squared)

# With condition
energy_readings = [1450, 1520, 1380, 2600, 1490, 2800]
high_readings = [r for r in energy_readings if r > 2000]
print("High readings:", high_readings)

# Transform and filter
costs = [r * 0.12 for r in energy_readings if r > 1500]
print("Costs for high readings:", costs)
```

## 9. Working with Files

Reading and writing files is essential for data science.

```python
# Writing to a file
energy_data = [
    "timestamp,energy_kwh",
    "2024-01-01 00:00,1450",
    "2024-01-01 01:00,1520",
    "2024-01-01 02:00,1380"
]

with open("sample_energy.csv", "w") as file:
    for line in energy_data:
        file.write(line + "\n")

print("✅ File written successfully!")

# Reading from a file
with open("sample_energy.csv", "r") as file:
    lines = file.readlines()
    print("\nFile contents:")
    for line in lines:
        print(" ", line.strip())
```

## 🎯 Practice Challenge

Let's combine everything you've learned!

```python
# Challenge: Energy Monitoring System

# Sample data: building energy readings for one week (hourly)
weekly_readings = {
    "Monday": [1400, 1450, 1380, 1500, 1600, 1550],
    "Tuesday": [1420, 1480, 1390, 1510, 1620, 1560],
    "Wednesday": [1380, 1430, 1350, 2800, 1580, 1520],  # Note the spike!
    "Thursday": [1410, 1460, 1370, 1490, 1590, 1540],
    "Friday": [1390, 1440, 1360, 1480, 800, 1530],  # Note the drop!
}

# Task 1: Calculate average energy per day
print("📊 Daily Averages:")
for day, readings in weekly_readings.items():
    avg = sum(readings) / len(readings)
    print(f"  {day}: {avg:.0f} kWh")

# Task 2: Find anomalies (readings > 2000 or < 1000)
print("\n⚠️  Anomalies Detected:")
for day, readings in weekly_readings.items():
    for i, reading in enumerate(readings):
        if reading > 2000 or reading < 1000:
            print(f"  {day} Hour {i}: {reading} kWh")

# Task 3: Calculate total weekly consumption
total_consumption = sum([sum(readings) for readings in weekly_readings.values()])
print(f"\n💡 Total Weekly Consumption: {total_consumption} kWh")

# Task 4: Estimate cost (at $0.12 per kWh)
total_cost = total_consumption * 0.12
print(f"💰 Estimated Weekly Cost: ${total_cost:.2f}")
```
