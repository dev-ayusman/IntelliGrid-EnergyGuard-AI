## 1. NumPy - Numerical Python

### Why NumPy?
- **Fast**: 50-100x faster than Python lists for numerical operations
- **Powerful**: Supports multi-dimensional arrays and matrices
- **Foundation**: Pandas and scikit-learn are built on NumPy

````python
import numpy as np

# Creating arrays
energy_list = [1450, 1520, 1380, 1600, 1490]
energy_array = np.array(energy_list)

print("Python list:", energy_list)
print("NumPy array:", energy_array)
print("Type:", type(energy_array))
print("Shape:", energy_array.shape)
print("Data type:", energy_array.dtype)
````

### NumPy Operations - Vectorization Magic! ✨

````python
# Mathematical operations (element-wise)
energy = np.array([1450, 1520, 1380, 1600, 1490])

# Convert kWh to cost (at $0.12 per kWh)
cost = energy * 0.12
print("Costs:", cost)

# Add 100 to each value
increased = energy + 100
print("Increased:", increased)

# Statistical operations
print("\nStatistics:")
print("  Mean:", np.mean(energy))
print("  Median:", np.median(energy))
print("  Std Dev:", np.std(energy))
print("  Min:", np.min(energy))
print("  Max:", np.max(energy))
print("  Sum:", np.sum(energy))
````

### 2D Arrays - Like Excel Spreadsheets

````python
# Energy readings for 3 buildings over 5 hours
readings = np.array([
    [1450, 1520, 1380, 1600, 1490],  # Building 1
    [1200, 1250, 1180, 1300, 1220],  # Building 2
    [1800, 1850, 1780, 1900, 1820]   # Building 3
])

print("Readings shape:", readings.shape)  # (3 buildings, 5 hours)
print("\nReadings:")
print(readings)

# Access individual building
print("\nBuilding 1 readings:", readings[0])
print("Building 2, Hour 3:", readings[1, 2])

# Operations on axes
print("\nAverage per building (row-wise):", np.mean(readings, axis=1))
print("Average per hour (column-wise):", np.mean(readings, axis=0))
print("Overall average:", np.mean(readings))
````

### Boolean Indexing - Finding Anomalies!

````python
energy = np.array([1450, 1520, 2800, 1600, 900, 1490])

# Create boolean mask
high_usage = energy > 2000
print("High usage mask:", high_usage)

# Filter values
anomalies = energy[high_usage]
print("Anomalies (>2000):", anomalies)

# Complex conditions
abnormal = (energy > 2000) | (energy < 1000)  # | means OR
print("Abnormal readings:", energy[abnormal])

# Count anomalies
print(f"\nFound {np.sum(abnormal)} anomalies out of {len(energy)} readings")
````

## 2. Pandas - Your Data Analysis Superpower! 🐼

### Why Pandas?
- **DataFrames**: Like Excel, but programmable
- **Powerful**: Filter, group, merge, aggregate data easily
- **Industry Standard**: Used everywhere in data science

````python
import pandas as pd
import numpy as np

# Creating a DataFrame from a dictionary
data = {
    'building_id': ['A', 'B', 'C', 'D', 'E'],
    'energy_kwh': [1450, 1520, 1380, 1600, 1490],
    'temperature': [22.5, 23.1, 21.8, 24.5, 22.9],
    'occupancy': [120, 95, 140, 88, 105]
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print("\nShape:", df.shape)
print("Columns:", df.columns.tolist())
````

### Exploring Data - First Look

````python
# Create a larger sample dataset
np.random.seed(42)
n_samples = 100

sample_data = {
    'timestamp': pd.date_range('2024-01-01', periods=n_samples, freq='H'),
    'energy_kwh': np.random.normal(1500, 200, n_samples),
    'temperature': np.random.normal(22, 3, n_samples),
    'humidity': np.random.normal(60, 10, n_samples)
}

df = pd.DataFrame(sample_data)

# Essential methods
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nDataFrame info:")
print(df.info())

print("\nStatistical summary:")
print(df.describe())
````

### Selecting Data - Rows and Columns

````python
# Select single column (returns Series)
energy = df['energy_kwh']
print("Energy column (first 5):")
print(energy.head())

# Select multiple columns (returns DataFrame)
subset = df[['energy_kwh', 'temperature']]
print("\nSubset (first 5):")
print(subset.head())

# Select rows by index
print("\nRows 10-15:")
print(df.iloc[10:15])

# Select by condition
high_energy = df[df['energy_kwh'] > 1700]
print(f"\nHigh energy readings: {len(high_energy)} out of {len(df)}")
print(high_energy.head())
````

### Adding & Modifying Columns

````python
# Add new column
df['cost_usd'] = df['energy_kwh'] * 0.12

# Create derived column
df['hour'] = df['timestamp'].dt.hour
df['is_peak_hour'] = (df['hour'] >= 9) & (df['hour'] <= 17)

# Conditional column
df['energy_level'] = pd.cut(
    df['energy_kwh'],
    bins=[0, 1400, 1600, np.inf],
    labels=['Low', 'Normal', 'High']
)

print("Enhanced DataFrame:")
print(df[['timestamp', 'energy_kwh', 'cost_usd', 'energy_level']].head(10))
````

### Handling Missing Data

````python
# Introduce some missing values
df_with_missing = df.copy()
df_with_missing.loc[5:10, 'energy_kwh'] = np.nan
df_with_missing.loc[15:18, 'temperature'] = np.nan

print("Missing values per column:")
print(df_with_missing.isnull().sum())

# Check if any missing values
print(f"\nAny missing values? {df_with_missing.isnull().any().any()}")

# Drop rows with missing values
df_dropped = df_with_missing.dropna()
print(f"\nAfter dropna: {len(df_dropped)} rows (from {len(df_with_missing)})")

# Fill missing values
df_filled = df_with_missing.copy()
df_filled['energy_kwh'] = df_filled['energy_kwh'].fillna(df_filled['energy_kwh'].mean())
df_filled['temperature'] = df_filled['temperature'].fillna(method='ffill')  # Forward fill

print("\nAfter filling:")
print(df_filled.isnull().sum())
````

### Grouping and Aggregation

````python
# Add day of week
df['day_of_week'] = df['timestamp'].dt.day_name()

# Group by day and calculate statistics
daily_stats = df.groupby('day_of_week')['energy_kwh'].agg([
    ('count', 'count'),
    ('mean', 'mean'),
    ('min', 'min'),
    ('max', 'max'),
    ('std', 'std')
])

print("Daily statistics:")
print(daily_stats)

# Group by energy level
level_stats = df.groupby('energy_level')['cost_usd'].agg(['mean', 'sum'])
print("\nCost by energy level:")
print(level_stats)
````

### Sorting and Ranking

````python
# Sort by energy consumption
sorted_df = df.sort_values('energy_kwh', ascending=False)
print("Top 5 highest energy readings:")
print(sorted_df[['timestamp', 'energy_kwh', 'temperature']].head())

# Multiple column sorting
sorted_multi = df.sort_values(['energy_level', 'energy_kwh'], ascending=[True, False])
print("\nSorted by level then energy:")
print(sorted_multi[['timestamp', 'energy_kwh', 'energy_level']].head(10))
````

## 3. Data Visualization

### Matplotlib - Basic Plotting

````python
import matplotlib.pyplot as plt

# Line plot
plt.figure(figsize=(12, 4))
plt.plot(df['timestamp'], df['energy_kwh'], linewidth=1, alpha=0.7)
plt.axhline(y=df['energy_kwh'].mean(), color='r', linestyle='--', label='Average')
plt.xlabel('Time')
plt.ylabel('Energy (kWh)')
plt.title('Energy Consumption Over Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Average energy: {df['energy_kwh'].mean():.2f} kWh")
````

### Multiple Subplots

````python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Energy over time
axes[0, 0].plot(df['timestamp'], df['energy_kwh'], color='steelblue')
axes[0, 0].set_title('Energy Consumption')
axes[0, 0].set_ylabel('kWh')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Temperature over time
axes[0, 1].plot(df['timestamp'], df['temperature'], color='coral')
axes[0, 1].set_title('Temperature')
axes[0, 1].set_ylabel('°C')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Energy distribution (histogram)
axes[1, 0].hist(df['energy_kwh'], bins=30, color='lightgreen', edgecolor='black')
axes[1, 0].set_title('Energy Distribution')
axes[1, 0].set_xlabel('kWh')
axes[1, 0].set_ylabel('Frequency')

# Plot 4: Scatter plot
axes[1, 1].scatter(df['temperature'], df['energy_kwh'], alpha=0.5, color='purple')
axes[1, 1].set_title('Energy vs Temperature')
axes[1, 1].set_xlabel('Temperature (°C)')
axes[1, 1].set_ylabel('Energy (kWh)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### Seaborn - Statistical Visualization

````python
import seaborn as sns

# Set style
sns.set_style('whitegrid')

# Box plot by energy level
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='energy_level', y='energy_kwh', palette='Set2')
plt.title('Energy Distribution by Level')
plt.ylabel('Energy (kWh)')
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation = df[['energy_kwh', 'temperature', 'humidity']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

print("\nCorrelation values:")
print(correlation)
````

## 4. Real-World Example: Exploratory Data Analysis

````python
# Create a realistic energy dataset
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=168, freq='H')  # 1 week hourly

# Simulate hourly patterns
hours = dates.hour
base_consumption = 1200
hour_effect = np.where((hours >= 9) & (hours <= 17), 400, 0)  # Peak hours
noise = np.random.normal(0, 100, len(dates))

energy_data = pd.DataFrame({
    'timestamp': dates,
    'energy_kwh': base_consumption + hour_effect + noise,
    'temperature': 20 + 5 * np.sin(2 * np.pi * hours / 24) + np.random.normal(0, 2, len(dates)),
    'humidity': 60 + np.random.normal(0, 10, len(dates))
})

# Add some anomalies
energy_data.loc[50:52, 'energy_kwh'] *= 1.8  # Spike
energy_data.loc[100:103, 'energy_kwh'] *= 0.3  # Drop

print("Dataset shape:", energy_data.shape)
print("\nFirst few rows:")
print(energy_data.head(10))
print("\nBasic statistics:")
print(energy_data.describe())
````

````python
# Detect anomalies using statistical method
mean = energy_data['energy_kwh'].mean()
std = energy_data['energy_kwh'].std()
threshold = 2

energy_data['z_score'] = np.abs((energy_data['energy_kwh'] - mean) / std)
energy_data['is_anomaly'] = energy_data['z_score'] > threshold

anomalies = energy_data[energy_data['is_anomaly']]
print(f"Detected {len(anomalies)} anomalies\n")
print(anomalies[['timestamp', 'energy_kwh', 'z_score']])

# Visualize
plt.figure(figsize=(14, 6))
plt.plot(energy_data['timestamp'], energy_data['energy_kwh'], label='Normal', linewidth=1)
plt.scatter(anomalies['timestamp'], anomalies['energy_kwh'], 
           color='red', s=100, label='Anomaly', zorder=5)
plt.axhline(y=mean, color='green', linestyle='--', label='Mean', alpha=0.7)
plt.axhline(y=mean + 2*std, color='orange', linestyle=':', label='±2σ threshold', alpha=0.7)
plt.axhline(y=mean - 2*std, color='orange', linestyle=':', alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Energy (kWh)')
plt.title('Energy Consumption with Anomaly Detection')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
